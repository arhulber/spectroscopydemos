```python
import numpy as np
import matplotlib.pyplot as plt
import os


os.makedirs('data_3peak', exist_ok=True)  # Create the directory if it doesn't exist

x = np.linspace(0, 100, 1000) # Generate 1000 points between 0 and 100

def gaussian(x, amp, cen, wid):
    return amp*np.exp(-((x - cen) ** 2) / (2 * wid ** 2))

#define true absorption peaks before noise
y_true = (gaussian(x, 5, 20, 3) + gaussian(x, 3, 50, 5) + gaussian(x, 4, 80, 2))
noise = np.random.normal(0, 0.2, x.shape) # Generate noise

y_data = y_true + noise

plt.figure(figsize = (10, 6))
plt.plot(x, y_data, label = 'Noisy Data', color = 'blue')
plt.plot(x, y_true, label = 'True Peaks', color = 'orange')
plt.xlabel("wavelength_nm")
plt.ylabel('absorbance')
plt.legend()
plt.title("Synthetic Spectrum with 3 Peaks")
plt.show()
```
